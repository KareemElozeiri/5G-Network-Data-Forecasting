```python
import pandas as pd
```

```python
df = pd.read_csv("../data/ue-lte-network-traffic-stats.csv")
```

```python
df.head()
```

```python
day_sample_size = 3600
target_columns = ["UE1-CQI", "UE2-CQI", "UE3-CQI"]
```

```python
for col in target_columns:
    for i in range(13):
        df[col][i*day_sample_size:(i+1)*day_sample_size].reset_index().to_csv(f"../data/{col}/{col}_{i+1}.csv")
```
